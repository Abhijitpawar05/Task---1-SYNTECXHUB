```python
#Load liblary
import numpy as np
```

```python
#Array creation , indexing and slicing
arr1 = np.array([10,20,30,40,50])
print(arr1)
```

```python
# To 2D array
arr2 = np.array([[1,2,3],
                 [4,5,6]])
print(arr2)
```

```python
#Array Indexing 
print(arr1[2])
```

```python
#To print 2D Indexing
print(arr2[0,1])
```

```python
#Array Slicing
print(arr1[1:4])
```

```python
# Mathmatical operations
print(arr1 + 5)
```

```python
print(arr1 * 2)
```

```python
print(np.sqrt(arr1))
```

```python
#Axis-wise opertions 
print(np.sum(arr2, axis=0)) #column wiae sum
print(np.sum(arr2, axis=1)) #Row wiae sum
```

```python
#Statistical Operation
print(np.mean(arr1)) #Find Mean
```

```python
print(np.median(arr1)) #Find Median
```

```python
print(np.std(arr1)) #Find Standard deviation 
```

```python
#To find out Max and min
print(np.max(arr1))
print(np.min(arr1))
```

```python
#Reshaping and Brodcasting

###Reshaping
a = np.array([1,2,3,4])
b = a.reshape(2,2)
print(b)
```

```python
#Brodcasting 
c = np.array([1,2,3])
d = np.array([[1],[2],[3]])
print(c + d)
```

```python
#Save and Load Numpy Array
np.save("myarray.npy", arr1)

loaded = np.load("myarray.npy")
print(loaded)
```

```python
#Compare performance Python list and Numpy list

### Python List
import time
list_data = list(range(1000000))
start = time.time()
result = [x * 2 for x in list_data]
print("List Time:", time.time() - start)
```

```python
# NumPy array
array_data = np.arange(1000000)
start = time.time()
result = array_data * 2
print("NumPy Time:", time.time() - start)
```
